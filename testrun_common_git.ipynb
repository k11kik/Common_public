## upload

codeを実行すると, 「Total files: XX. Proceed with upload? [Y/n]:」と表示されるので, Enterもしくは"y"を押してEnter

* jupyter notebookの場合は, 上方のコマンドパレットに上記メッセージが表示されるので, 同じくEnterもしくは"y"を押してEnter

-> uploadが開始される

### upload_github()関数

| **Parameter** | **Type** | **Default** | **Notes** |
| :--- | :--- | :--- | :--- |
| owner | str | <span style="color: yellow;">Required</span> | githubリポジトリのオーナー名 |
| repo | str | <span style="color: yellow;">Required</span> | githubリポジトリ名 |
| branch | str | <span style="color: yellow;">Required</span> | ブランチ名. とりあえずは'main'でok.|
| local_path | str | <span style="color: yellow;">Required</span> | アップロードするローカルのファイルまたはディレクトリパス |
| remote_path | str | <span style="color: yellow;">Required</span> | GitHub上のアップロード先パス |
| github_token | str | None | GitHub Personal Access Token |
| is_dir | bool | False | アップロード対象がディレクトリである場合はTrue. ファイルの場合はFalse |
| commit_message | str | 'Update via script' | GitHubへのコミットメッセージ |
| confirm | bool | True | アップロード実行前に確認を求めるかどうか |
| info | bool | True | 進捗情報などの主要情報を表示するかどうか |
| subinfo | bool | False | 詳細なサブ情報を表示するかどうか |
| extensions | list[str] | None | uploadするファイルの拡張子. 何も指定しなければpythonファイルが指定される. 複数指定したい場合は['.py', '.ipynb']などのようにlistで指定.  |



directoryをuploadする場合 -> is_dir=True

```python
git.upload_github(
    owner='k11kik',
    repo='Common_public',
    branch='main',
    local_path='common',
    remote_path='common',
    is_dir=True,
    github_token=github_token
)
```

file単体をuploadする場合 -> is_dir=False

```python
git.upload_github(
    owner='k11kik',
    repo='Common_public',
    branch='main',
    local_path='testrun_common_git.ipynb',
    remote_path='',
    is_dir=False,
    github_token=github_token,
    extensions='.ipynb'
)
```
