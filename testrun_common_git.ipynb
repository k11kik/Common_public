# Test Run: Common/git

目的: 
pythonを用いて, ファイルをgithubからdownloadしたり, githubへuploadする. 

こんな人へ:
* コードのback upを取りたい. 
* VScodeとGitHubの連携が負担

---

準備: 
1. pythonファイルdownload
    1. https://github.com/k11kik/Common_public/tree/main へアクセス
    2. Code > Download ZIP
    3. zipファイルを解凍
2. (a) path設定 [Project folderがない場合]
    1. localにフォルダ「Common」を作成
    2. 「Common」中にdownloadしたzipファイル「Common_public-main」の中身を移動
    3. 「Common/common/git」, 「Common/testrun_common_git.ipynb」のような構造になっていればok.

    (b) path設定 [Project folderがある場合]
    1. Project folder中にdownloadしたzipファイル「Common_public-main」の中身を移動
    2. 「ProjectName/common/git」, 「ProjectName/testrun_common_git.ipynb」のような構造になっていればok.
3. GitHub設定
    1. GitHubアカウントを作成
    2. リポジトリを作成 (リポジトリ名: 「Common」)
        * 公開する予定がなければ「private」をお勧めします
    3. github tokenを取得
        * profil > Settings > Developer settings > Personal access tokens > Tokens (classic) > Generate new token
        * Select scopes > repoにチェックを入れる
        * 作成されたtokenをコピペ

## upload

codeを実行すると, 「Total files: XX. Proceed with upload? [Y/n]:」と表示されるので, Enterもしくは"y"を押してEnter

* jupyter notebookの場合は, 上方のコマンドパレットに上記メッセージが表示されるので, 同じくEnterもしくは"y"を押してEnter

-> uploadが開始される

### upload()関数

| **Parameter** | **Type** | **Default** | **Notes** |
| :--- | :--- | :--- | :--- |
| owner | str | <span style="color: yellow;">Required</span> | githubリポジトリのオーナー名 |
| repo | str | <span style="color: yellow;">Required</span> | githubリポジトリ名 |
| local_path | str | <span style="color: yellow;">Required</span> | アップロードするローカルのファイルまたはディレクトリパス |
| github_token | str | <span style="color: yellow;">Required</span> | GitHub Personal Access Token |
| branch | str | 'main' | ブランチ名. とりあえずは'main'でok.|
| extensions | list[str] | None | uploadするファイルの拡張子. 何も指定しなければpythonファイルが指定される. 複数指定したい場合は['.py', '.ipynb']などのようにlistで指定.  |
| remote_path | str | None | GitHub上のアップロード先パス |
| commit_message | str | 'Update via script' | GitHubへのコミットメッセージ |
| confirm | bool | True | アップロード実行前に確認を求めるかどうか |



directoryをuploadする場合 -> is_dir=True

```python
git.upload(
    owner='k11kik',
    repo='Common',
    # repo='Common_public',
    local_path='common',
    github_token=github_token
)
```

```python
git.upload(
    owner='k11kik',
    repo='Common',
    # repo='Common_public',
    local_path='testrun_common_git.ipynb',
    github_token=github_token
)
```
