# Test Run: Common/cdf

cdf fileは, txtやcsvと同様, データを格納して保存が可能です. txt, csv fileでは1次元配列を保存することはできるが, 多次元配列を保存することはできません. 対して, cdf fileでは1次元配列でも多次元配列でもnumpy配列をそのまま保存可能で, とても便利です. 

cdf moduleには, データ保存およびデータ読み出し機能があります. 

目的: 
CDF fileの読み出し/書き出しを簡単に行う

こんな人へ:
* CDFって何？
* numpy.ndarrayのような多次元配列をそのまま保存したい
* txtやcsvへのデータ格納が手間

```python
import numpy as np
from common import cdf
```

sample data

```python
arr_1d = np.linspace(0, 1, 100)
arr_2d = np.zeros((10, 10))
str_data = 'sample data'
print(f'{arr_1d=}')
print(f'{arr_2d=}')
```

CDFでは, データ名とデータを結びつけて保存します. 辞書型配列(dict)と同じなので, dict配列を作ります. 

```python
dict_data = {
    'arr_1d': arr_1d,
    'arr_2d': arr_2d,
    'str_data': str_data,
}
print(dict_data)
```

## CDF fileへ書き出し: dict -> CDF

dict dataをCDF fileとして保存します. 

cdf.dict_to_cdffile()を用います. 

```python
savecdf = 'test.cdf' # 保存する先のCDF file path
cdf.dict_to_cdffile(dict_data, savecdf)
```

これで, localに「test.cdf」が生成されたはずです. 次は生成されたCDF fileを読み込みます

## CDF fileの読み込み: CDF -> dict

cdf.cdffile_to_dict()を用います. 

```python
cdf_filepath = 'test.cdf'
dict_data = cdf.cdffile_to_dict(cdf_filepath)
print(f'{dict_data=}')
```

正しくdict dataが読み込めていればokです. それぞれの変数データを取り出したければ, 通常のdict配列と同様の処理をすれば良いです. 

```python
arr_1d = dict_data['arr_1d']
arr_2d = dict_data['arr_2d']
str_data = dict_data['str_data']
print(f'{arr_1d=}')
print(f'{arr_2d=}')
print(f'{str_data=}')
```

cdf.info()とすれば, cdf fileの中身の情報を表示できます

```python
cdf.info(cdf_filepath)
```
