# Test Run: Common/cdf

cdf fileは, txtやcsvと同様, データを格納して保存が可能です. txt, csv fileでは1次元配列を保存することはできるが, 多次元配列を保存することはできません. 対して, cdf fileでは1次元配列でも多次元配列でもnumpy配列をそのまま保存可能で, とても便利です. 

cdf moduleには, データ保存およびデータ読み出し機能があります. 

目的: 
CDF fileの読み出し/書き出しを簡単に行う

こんな人へ:
* CDFって何？
* numpy.ndarrayのような多次元配列をそのまま保存したい
* txtやcsvへのデータ格納が手間

ちなみに, Autoplotという公開ツールを使えば, cdf fileの中身を簡単に可視化することができます.

---

準備: 
1. pythonファイルdownload
    1. https://github.com/k11kik/Common_public/tree/main へアクセス
    2. Code > Download ZIP
    3. zipファイルを解凍
2. (a) path設定 [Project folderがない場合]
    1. localにフォルダ「Common」を作成
    2. 「Common」中にdownloadしたzipファイル「Common_public-main」の中身を移動
    3. 「Common/common/cdf」, 「Common/testrun_common_cdf.ipynb」のような構造になっていればok.

    (b) path設定 [Project folderがある場合]
    1. Project folder中にdownloadしたzipファイル「Common_public-main」の中身を移動
    2. 「ProjectName/common/cdf」, 「ProjectName/testrun_common_cdf.ipynb」のような構造になっていればok.

```python
import numpy as np
import matplotlib.pyplot as plt
from common_public import cdf
```

# Save data as CDF file: dict -> CDF

```python
# sample data
arr_1d = np.linspace(0, 1, 100)
arr_2d = np.zeros((10, 10))
str_data = 'sample data'
print(f'{arr_1d=}')
print(f'{arr_2d=}')
```

CDFでは, データ名とデータを結びつけて保存します. 辞書型配列(dict)と同じなので, dict配列を作ります. 

```python
dict_data = {
    'arr_1d': arr_1d,
    'arr_2d': arr_2d,
    'str_data': str_data,
}
print(dict_data)
```

```python
# データに説明を付す場合
dict_data = {
    'arr_1d': {'data': arr_1d, 'unit': 'X', 'desc': 'array data (1D)'},
    'arr_2d': {'data': arr_2d, 'unit': 'Y', 'desc': 'array data (1D)'},
    'str_data': {'data': str_data, 'unit': 'Z', 'desc': 'string data (str)'},
}
```

dict dataをCDF fileとして保存します. 

cdf.dict_to_cdffile()を用います. 

```python
savecdf = 'test.cdf' # 保存する先のCDF file path
cdf.dict_to_cdffile(dict_data, savecdf)
```

これで, localに「test.cdf」が生成されたはずです. 次は生成されたCDF fileを読み込みます

## Read CDF file: CDF -> dict

まずは中身の概要をcdf.info()で表示してみます. 

```python
cdf_filepath = 'test.cdf'
cdf.info(cdf_filepath)
```

infoが表示されていれば正しく読み込めています. では, cdf.cdffile_to_dict()を用いてデータを読み込んでみましょう. 

```python
cdf_filepath = 'test.cdf'
dict_data = cdf.cdffile_to_dict(cdf_filepath)
print(f'{dict_data=}')
```

正しくdict dataが読み込めていればokです. それぞれの変数データを取り出したければ, 通常のdict配列と同様の処理をすれば良いです. 

```python
arr_1d = dict_data['arr_1d']
arr_2d = dict_data['arr_2d']
str_data = dict_data['str_data']
print(f'{arr_1d=}')
print(f'{arr_2d=}')
print(f'{str_data=}')
```

# Example: Arase Data

https://ergsc.isee.nagoya-u.ac.jp/data/ergsc/satellite/erg/mgf/l2/8sec/2020/01/

から, 「erg_mgf_l2_8sec_20200101_v04.06.cdf」をDownload -> このfileと同じディレクトリに保存

```python
cdf_filepath = 'erg_mgf_l2_8sec_20200101_v04.06.cdf'
# info
cdf.info(cdf_filepath)
```

```python
# cdf -> dict
dict_data = cdf.cdffile_to_dict(cdf_filepath)
epoch_8sec = dict_data['epoch_8sec']
mag_8sec_dsi = dict_data['mag_8sec_dsi']

# plot
fig, axes = plt.subplots(figsize=(12, 6))
axes.plot(epoch_8sec, mag_8sec_dsi)
```

```python
# invalid data 除去
mag_8sec_dsi = np.where(mag_8sec_dsi == -1e31, np.nan, mag_8sec_dsi)

# plot
fig, axes = plt.subplots(figsize=(12, 6))
axes.plot(epoch_8sec, mag_8sec_dsi, label=['Bx', 'By', 'Bz'])
axes.legend()

```

```python
# 他のデータも同様に可視化
mag_8sec_gsm = dict_data['mag_8sec_gsm']
mag_8sec_gse = dict_data['mag_8sec_gse']
mag_8sec_sm = dict_data['mag_8sec_sm']

# Exclude invalid data
mag_8sec_gsm = np.where(mag_8sec_gsm == -1e31, np.nan, mag_8sec_gsm)
mag_8sec_gse = np.where(mag_8sec_gse == -1e31, np.nan, mag_8sec_gse)
mag_8sec_sm = np.where(mag_8sec_sm == -1e31, np.nan, mag_8sec_sm)

# plot
fig, axes = plt.subplots(4, 1, figsize=(12, 10))
axes[0].plot(epoch_8sec, mag_8sec_dsi, label=['Bx', 'By', 'Bz'])
axes[1].plot(epoch_8sec, mag_8sec_gsm, label=['Bx', 'By', 'Bz'])
axes[2].plot(epoch_8sec, mag_8sec_gse, label=['Bx', 'By', 'Bz'])
axes[3].plot(epoch_8sec, mag_8sec_sm, label=['Bx', 'By', 'Bz'])
axes[0].legend(loc='upper right')
axes[1].legend(loc='upper right')
axes[2].legend(loc='upper right')
axes[3].legend(loc='upper right')

```
